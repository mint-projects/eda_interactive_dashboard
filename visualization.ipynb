# Initial notebook to understand and visualize the data


```python
from dotenv import load_dotenv
from kaggle.api.kaggle_api_extended import KaggleApi
import os.path

load_dotenv()

api = KaggleApi()
api.authenticate()

dataset_slug = 'algozee/smartphone-usage-and-addiction-analysis-dataset'

fileAlreadyExists = os.path.isfile("data/Smartphone_Usage_And_Addiction_Analysis_7500_Rows.csv")

if fileAlreadyExists:
    print("Data already fetched")
else:
    print(f"Downloading {dataset_slug}...")
    api.dataset_download_files(dataset_slug, path='data/', unzip=True)
    print("Files downloaded and ready to use")
```
