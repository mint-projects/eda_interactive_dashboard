```python
import os
from dotenv import load_dotenv
from kaggle.api.kaggle_api_extended import KaggleApi

# 1. Wczytanie kluczy z .env
load_dotenv()

# 2. Logowanie do API
api = KaggleApi()
api.authenticate()

# 3. Pobieranie danych
# Zmień 'piterfm/2022-ukraine-russian-war' na slug swojego zbioru
dataset_slug = 'algozee/smartphone-usage-and-addiction-analysis-dataset'

print(f"Pobieranie {dataset_slug}...")
api.dataset_download_files(dataset_slug, path='.', unzip=True)
print("Gotowe! Pliki są w folderze projektu.")
```
