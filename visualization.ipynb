# Initial notebook to understand and visualize the data


```python
from dotenv import load_dotenv
from kaggle.api.kaggle_api_extended import KaggleApi
import os.path

load_dotenv()

api = KaggleApi()
api.authenticate()

dataset_slug = 'algozee/smartphone-usage-and-addiction-analysis-dataset'

fileAlreadyExists = os.path.isfile("data/Smartphone_Usage_And_Addiction_Analysis_7500_Rows.csv")

if fileAlreadyExists:
    print("Data already fetched")
else:
    print(f"Downloading {dataset_slug}...")
    api.dataset_download_files(dataset_slug, path='data/', unzip=True)
    print("Files downloaded and ready to use")
```

## Initial data lookup


```python
import pandas as pd
from IPython.display import display

data = pd.read_csv("data/Smartphone_Usage_And_Addiction_Analysis_7500_Rows.csv")

display(data.head())
```

```python
data.describe()
```

```python
data.isna().sum()
```

```python
print(data["addiction_level"].unique())
```

### Investigating, why addiction level feature has missing values

```python
missing = data[data["addiction_level"].isnull()].copy()
not_missing = data[data["addiction_level"].notnull()].copy()

print("Mean values for not missing data:")
print(not_missing.mean(numeric_only=True))

print("\nMean values for missing values in the addiciton level column:")
print(missing.mean(numeric_only=True))
```

```python
display(missing.head(30))
```

### It turns out, that the NaN value in addiction level feature probably just indicates that a person is not addicted, so I will just impute the NaN values with a string that is more informative

```python
data["addiction_level"] = data["addiction_level"].fillna("Not addicted")
data.isna().sum()
```

```python
data = data.drop(columns=["transaction_id", "user_id"])
```

## Saving the data for visualization


```python
data.to_csv("data/clean_data.csv")
```
